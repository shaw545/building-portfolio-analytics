Clean text columns

```python
text_cols = ["building_id", "building_name", "city", "state", "property_type"]
for c in text_cols:
    df[c] = df[c].astype(str).str.strip()

df[text_cols].head()
```

```python

num_cols = [
    "year_built","floors","total_units","total_sqft","occupied_units",
    "avg_monthly_rent","annual_other_income","annual_opex",
    "annual_property_tax","annual_insurance","annual_maintenance",
    "market_value","lat","lon"
]
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df[num_cols].head()
```

```python
non_negative = [
    "floors","total_units","total_sqft","occupied_units","avg_monthly_rent",
    "annual_other_income","annual_opex","annual_property_tax",
    "annual_insurance","annual_maintenance","market_value"
]
for c in non_negative:
    df.loc[df[c] < 0, c] = np.nan
```

```python
mask = (df["total_units"] > 0) & (df["occupied_units"] > df["total_units"])
df.loc[mask, "occupied_units"] = df.loc[mask, "total_units"]
```

Creating annual rent (key assumption)
Assumption:

Multifamily: rent is $/unit/month

Office/Retail: rent is $/sqft/month


```python
df["gross_rent_annual"] = np.where(
    df["property_type"].str.lower().eq("multifamily"),
    df["occupied_units"] * df["avg_monthly_rent"] * 12,
    df["total_sqft"] * df["avg_monthly_rent"] * 12
)

df[["building_name","property_type","gross_rent_annual"]].head(10)
```

Total income, total expenses, NOI

```python
df["gross_income_annual"] = df["gross_rent_annual"] + df["annual_other_income"]

df["total_expenses_annual"] = (
    df["annual_opex"]
    + df["annual_property_tax"]
    + df["annual_insurance"]
    + df["annual_maintenance"]
)

df["noi_annual"] = df["gross_income_annual"] - df["total_expenses_annual"]

df[["building_name","gross_income_annual","total_expenses_annual","noi_annual"]].head(10)
```

Cap Rate

```python
df["cap_rate"] = np.where(
    df["market_value"] > 0,
    df["noi_annual"] / df["market_value"],
    np.nan
)

df[["building_name","noi_annual","market_value","cap_rate"]].head(10)
```

Rent per sqft (monthly)

```python
df["rent_per_sqft_month"] = np.where(
    df["total_sqft"] > 0,
    np.where(
        df["property_type"].str.lower().eq("multifamily"),
        (df["occupied_units"] * df["avg_monthly_rent"]) / df["total_sqft"],
        df["avg_monthly_rent"]
    ),
    np.nan
)

df[["building_name","property_type","rent_per_sqft_month",]].head(10)
```

Summary tables

By property type

```python
summary_by_type = (
    df.groupby("property_type")
      .agg(
          buildings=("building_id","count"),
          avg_occupancy=("occupancy_rate","mean"),
          total_noi=("noi_annual","sum"),
          avg_cap_rate=("cap_rate","mean"),
          
      )
      .reset_index()
      .sort_values("total_noi", ascending=False)
)

summary_by_type
```

By city/state

```python
summary_by_city = (
    df.groupby(["city","state"])
      .agg(
          buildings=("building_id","count"),
          total_noi=("noi_annual","sum"),
          avg_cap_rate=("cap_rate","mean"),
          avg_occupancy=("occupancy_rate","mean")
      )
      .reset_index()
      .sort_values("total_noi", ascending=False)
)

summary_by_city
```

Ranking buildings (Top performers + risk list)

```python
top10_noi = df.sort_values("noi_annual", ascending=False).head(10)
top10_cap = df.dropna(subset=["cap_rate"]).sort_values("cap_rate", ascending=False).head(10)

top10_noi[["building_name","city","noi_annual","cap_rate"]]
```

```python

plt.figure()
plt.barh(top10_noi["building_name"], top10_noi["noi_annual"])
plt.gca().invert_yaxis()
plt.title("Top 10 Buildings by NOI (Annual)")
plt.xlabel("NOI ($)")
plt.tight_layout()
plt.show()
```
